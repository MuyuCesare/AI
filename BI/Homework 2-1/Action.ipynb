```python
import pandas as pd 
import numpy as np
from sklearn import preprocessing
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
#数据加载
data = pd.read_csv('./Market_Basket_Optimisation.csv', header=None)

print(data.shape)
```

```python
data.head()
```

```python
#把数据存放到Transactions中
transactions=[]
for i in range(0, data.shape[0]):
    temp=[]
    for j in range(0, data.shape[1]):
        if str(data.values[i, j]) != 'nan':
            temp.append(str(data.values[i, j]))
    #print(temp)
    transactions.append(temp)
```

```python
all_transactions=['|'.join(i) for i in transactions]
```

```python
all_transactions=pd.DataFrame(all_transactions,columns=['transactions'])
```

```python
transactions_hot_encoded = all_transactions.drop('transactions',1).join(all_transactions.transactions.str.get_dummies(sep='|'))
```

```python
itemsets = apriori(transactions_hot_encoded,use_colnames=True, min_support=0.02)
```

```python
# 按照支持度从大到小进行时候粗
itemsets = itemsets.sort_values(by="support" , ascending=False) 
print('-'*20, '频繁项集', '-'*20)
print(itemsets)
# 根据频繁项集计算关联规则，设置最小提升度为2
rules =  association_rules(itemsets, metric='lift', min_threshold=1.8)

# 按照提升度从大到小进行排序
rules = rules.sort_values(by="lift" , ascending=False) 
#rules.to_csv('./rules.csv')
print('-'*20, '关联规则', '-'*20)
print(rules)
```
